```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('./data.csv')
nulldata = data.isnull().sum() * 1./data.shape[0]
nulldata.plot(kind='bar')
plt.show()

dropcols = ['location_description', 'participant_name', 'participant_age_group', 'participant_relationship', 'date', 'address', 'incident_url', 'source_url', 'incident_url_fields_missing', 'incident_characteristics', 'notes', 'sources']
data = data.drop(dropcols, axis=1)
data["n_guns_involved"] = data["n_guns_involved"].fillna(-1).astype(int)
colnames = ['congressional_district', 'state_house_district', 'state_senate_district']
data[colnames] = data[colnames].astype(object)
parse_columns = ['gun_stolen', 'gun_type', 'incident_characteristics', 'participant_age', 'participant_age_group', 'participant_gender', 'participant_name', 'participant_status', 'participant_type']

# Make "incident_id" column the index
data.set_index('incident_id')
data['gun_stolen'] = data['gun_stolen'].str.replace('\|\|', ', ')
data['gun_stolen'] = data['gun_stolen'].str.replace('::', ': ')

data['participant_gender'] = data['participant_gender'].str.replace('\|\|', ', ')
data['participant_gender'] = data['participant_gender'].str.replace('::', ': ')

data['participant_age'] = data['participant_age'].str.replace('\|\|', ', ')
data['participant_age'] = data['participant_age'].str.replace('::', ': ')

data['participant_status'] = data['participant_status'].str.replace('\|\|', ', ')
data['participant_status'] = data['participant_status'].str.replace('::', ': ')

data['participant_type'] = data['participant_type'].str.replace('\|\|', ', ')
data['participant_type'] = data['participant_type'].str.replace('::', ': ')

data['gun_type'] = data['gun_type'].str.replace('\|\|', ', ')
data['gun_type'] = data['gun_type'].str.replace('::', ': ')

pd.set_option('display.max_columns', 500)

from nltk.tokenize import sent_tokenize, word_tokenize 
import warnings 
  
warnings.filterwarnings(action = 'ignore') 
  
import gensim 
from gensim.models import Word2Vec 
```

```python
def conv2num(data, colname):
    finaldata = []
    for state in data[:][colname]:
        finaldata.append([state])
    model1 = gensim.models.Word2Vec(finaldata, min_count = 1,size = 1, window = 1)
    dict1 = {}
    for state in model1.wv.vocab:
        dict1[state] = model1.wv[state][0]
    data[colname] = data[colname].map(dict1)
#     for i in range(len(data[:][colname])):
#         state1 = data.loc[data.index[i],colname]
#         data.loc[data.index[i],colname] = model1.wv[state1]
```

```python
def get_list_from_str(string):
    string_list = []
    strings = string.split(", ")
#     print(string)
    for single_string in strings:
#         print(single_string)
        if(": " in single_string):
            string_list.append(single_string.split(": ")[1])
    return string_list

def convAllLists(data):
    gun_stolen_list = []
    gun_type_list = []
    part_status_list = []
    part_type_list = []

    for row in range(data.shape[0]):
        gun_stolen = data.loc[data.index[row], "gun_stolen"]
        gun_type = data.loc[data.index[row], "gun_type"]
        part_age = data.loc[data.index[row], "participant_age"]
        part_gender = data.loc[data.index[row], "participant_gender"]
        part_status = data.loc[data.index[row], "participant_status"]
        part_type = data.loc[data.index[row], "participant_type"]
#         print(type(gun_stolen))
        if(not (part_age=='NaN' or isinstance(part_age, float))):
            ages = part_age.split(", ")
            avg_age = 0
            for age in ages:
                if(": " in age):
                    avg_age += int(age.split(": ")[1])
            avg_age = float(avg_age)/len(ages)
            data.loc[data.index[row], "participant_age"] = avg_age

        if(not (part_gender=='NaN' or isinstance(part_gender, float))):
            genders = part_gender.split(", ")
            avg_gender = 0
            for gender in genders:
                if(": " in gender):
                    isMale = (gender.split(": ")[1] == "Male")
                    if(isMale):
                        avg_gender += 1
            avg_gender = float(avg_gender)/len(genders)
            data.loc[data.index[row], "participant_gender"] = avg_gender

        if(not (gun_stolen=='NaN' or isinstance(gun_stolen, float))):
            gun_stolen_list.append(get_list_from_str(gun_stolen))
        if(not (gun_type=='NaN' or isinstance(gun_type, float))):
            gun_type_list.append(get_list_from_str(gun_type))
        if(not (part_status=='NaN' or isinstance(part_status, float))):
            part_status_list.append(get_list_from_str(part_status))
        if(not (part_type=='NaN' or isinstance(part_type, float))):
            part_type_list.append(get_list_from_str(part_type))

    model2 = gensim.models.Word2Vec(gun_stolen_list, min_count = 1,size = 1, window = 3)
    model3 = gensim.models.Word2Vec(gun_type_list, min_count = 1,size = 1, window = 3)
    model4 = gensim.models.Word2Vec(part_status_list, min_count = 1,size = 1, window = 3)
    model5 = gensim.models.Word2Vec(part_type_list, min_count = 1,size = 1, window = 3)

    dict2 = {}
    for state in model2.wv.vocab:
        dict2[state] = model2.wv[state][0]

    dict3 = {}
    for state in model3.wv.vocab:
        dict3[state] = model3.wv[state][0]

    dict4 = {}
    for state in model4.wv.vocab:
        dict4[state] = model4.wv[state][0]

    dict5 = {}
    for state in model5.wv.vocab:
        dict5[state] = model5.wv[state][0]

    colname = "gun_stolen"
    data[colname] = data[colname].map(dict1)
    colname = "gun_type"
    data[colname] = data[colname].map(dict1)
    colname = "participant_status"
    data[colname] = data[colname].map(dict1)
    colname = "participant_type"
    data[colname] = data[colname].map(dict1)
    
convAllLists(data)
```
